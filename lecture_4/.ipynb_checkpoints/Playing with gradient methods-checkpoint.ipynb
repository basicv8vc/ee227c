```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

np.random.seed(20171229)
from IPython.core.display import display, HTML
# display(HTML(open('custom.html').read()))

kwargs = {'linewidth' : 3.5}
font = {'weight' : 'normal', 'size'   : 24}
matplotlib.rc('font', **font)
```

```python
def loss_plot(ys, yscale='log'):
    plt.figure(figsize=(8, 8))
    plt.xlabel('Step')
    plt.ylabel('Loss')
    plt.yscale(yscale)
    plt.plot(range(len(ys)), ys, **kwargs)
```

# Projected gradient descent

```python
def gradient_descent(init, steps, grad, proj=lambda x: x):
    """ Projected gradient descent.
    Inputs:
        initial: starting point
        steps: list of scalar learning rates
        grad: function mapping points to gradients
        prof(optional): function mapping points to points
    Returns:
        List of all points computed by projected gradient descent.
    """
    xs = [init]
    for step in steps:
        xs.append(proj(xs[-1] - step * grad(xs[-1])))
    return xs
```

# Warm-up: Optimizing a quadratic

目标函数是:

$f(x) = \frac{1}{2}||x||^2$
 
梯度是: $\nabla f(x) = x$

```python
def quadratic(x):
    return 0.5*x.dot(x)
```

```python
def quadratic_gradient(x):
    return x
```

这个函数是1-smooth和1-strong convex，如果令学习率=1，一次迭代就收敛了

```python
x0 = np.random.normal(0, 1, (10))
```

```python
_, x1 = gradient_descent(x0, [1.0], quadratic_gradient)
```

```python
print(x0)
```

```python
print(x1)
```

```python
x1.all() == 0
```

如果我们不知道最优学习率

```python
xs = gradient_descent(x0, [5.5]*50, quadratic_gradient)
loss_plot([quadratic(x) for x in xs])
```

```python
xs = gradient_descent(x0, [.1]*50, quadratic_gradient)
loss_plot([quadratic(x) for x in xs])
```

# Constrained optimization

```python
U = np.linalg.qr(np.random.normal(0, 1, (1000, 100)))[0]
b = np.random.normal(0, 1, 1000)
```

```python
def proj(x):
    return b + U.dot(U.T).dot(x-b)
```

```python
x0 = np.random.normal(0, 1, (1000))
xs = gradient_descent(x0, [0.1]*50, quadratic_gradient, proj)
```

```python
x_opt = proj(0)
loss_plot([quadratic(x) for x in xs])
plt.plot(range(len(xs)), [quadratic(x_opt)] * len(xs),
        label = '$\\frac{1}{2}|\!|x_{\mathrm{opt}}|\!|^2$')
plt.legend()
```

# Least Squares   

最小化下面的损失函数:

$f(x)=\frac{1}{2m}\sum_{i=1}^m (a_{i}^T x-b_{j})^2=\frac{1}{2}||Ax-b||^2$

计算梯度:

$\nabla f(x) = A^T(Ax-b)$

$\nabla^2f(x) = A^TA$

目标函数是 $\beta$-smooth, 其中$\beta = \lambda_{max}(A^TA)$

$\alpha$-strongly convex, 其中$\alpha = \lambda_{min}(A^TA)$

```python
def least_squares(A, b, x):
    return (0.5) * np.linalg.norm(A.dot(x)-b)**2
```

```python
def least_squares_gradient(A, b, x):
    return A.T.dot(A.dot(x)-b)
```

```python
m, n = 1000, 100
A = np.random.normal(0, 1, (m, n))
x_opt = np.random.normal(0, 1, n)
noise = np.random.normal(0, 0.1, m)
b = A.dot(x_opt) + noise
objective = lambda x: least_squares(A, b, x)
gradient = lambda x: least_squares_gradient(A, b, x)
```

```python
x0 = np.random.normal(0, 1, n)
xs = gradient_descent(x0, [0.1]*100, gradient)
loss_plot([objective(x) for x in xs])

# plt.plot(range(len(xs)), [np.linalg.norm(noise)**2]*len(xs),
#         label='noise level', **kwargs)
# plt.plot(range(len(xs)), [least_squares(A,b,x_opt)]*len(xs),
#         label='optimal', **kwargs)
# _ = plt.legend()
```
